# Standard Deviation and Variance
Variance  =  Square(difference with the mean)/ N
Standard deviation  = Square root(variance)
Sample =  Square(difference with the mean)/ N-1

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

incomes = np.random.normal(200.0, 50.0, 100000)

plt.hist(incomes, 50)
plt.show()
```

```python
incomes.std()
```

```python
incomes.var()
```
